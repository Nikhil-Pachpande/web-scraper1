```python
# Import the Libraries
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Using BeautifulSoup as the Web Scraper
url = urlopen("https://en.wikipedia.org/wiki/Comparison_of_programming_languages")
bs = BeautifulSoup(url, "html.parser")
```

```python
# Scraping the tabular data from the url
table = bs.findAll("table", {"class":"wikitable"})[0]
rows = table.findAll("tr")
```

```python
# parsing the table and storing the data as csv file
with open("data.csv", "wt+", newline="") as f:
    writer = csv.writer(f)
    for i in rows:
        row = []
        for cell in i.findAll(["td", "th"]):
            row.append(cell.get_text())
        writer.writerow(row)
```
